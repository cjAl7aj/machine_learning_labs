```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
```

```python
path = "/content/covid19_statistics"
df = pd.read_csv(path)
df.head()
```

```python
df.info()
df.isna().sum()
df.describe()
```

```python
df_clean = df.copy()

numeric_cols = df_clean.select_dtypes(include=np.number).columns

for col in numeric_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mean())

df_clean.isna().sum()
```

```python
Q1 = df_clean[numeric_cols].quantile(0.25)
Q3 = df_clean[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df_no_outliers = df_clean[~((df_clean[numeric_cols] < lower) |
                             (df_clean[numeric_cols] > upper)).any(axis=1)]

print("Original shape:", df_clean.shape)
print("After removing outliers:", df_no_outliers.shape)
```

```python
scaler_minmax = MinMaxScaler()

df_minmax = df_no_outliers.copy()
df_minmax[numeric_cols] = scaler_minmax.fit_transform(df_minmax[numeric_cols])

df_minmax.head()
```

```python
scaler_standard = StandardScaler()

df_zscore = df_no_outliers.copy()
df_zscore[numeric_cols] = scaler_standard.fit_transform(df_zscore[numeric_cols])

df_zscore.head()
```

```python
X = df_zscore[numeric_cols]

pca = PCA()
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection")
plt.show()
```
